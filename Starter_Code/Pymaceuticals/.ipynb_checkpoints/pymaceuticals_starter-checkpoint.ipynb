### Analysis

- The mice that were treated with Capomulin or Ramicane show a final tumor volume smaller than mice treated with Infubinol or Ceftamin.
- The mouse weight and average tumor volume have a correlation of 0.84 in the Capomulin group, which indicates a strong and positive relationship.
- From the charts we can see that the distribution of male mice and female mice are almost equal.
- The Capomulin and Ramicane regimens had the best/strongest results compared to Infubinol or Ceftamin.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
merged_df = merged_df[["Mouse ID", "Timepoint", "Tumor Volume (mm3)", "Metastatic Sites", "Drug Regimen", "Sex", "Age_months", "Weight (g)"]]
merged_df.head()
```

```python
# Checking the number of mice.
total_mice = len(merged_df["Mouse ID"].unique())
total_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = merged_df.loc[merged_df.duplicated(subset=["Mouse ID", "Timepoint",]), "Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mouse_ID = merged_df.loc[merged_df["Mouse ID"] == "g989", :]
duplicate_mouse_ID
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_df = merged_df[merged_df["Mouse ID"].isin(duplicate_mice) ==False]
clean_mouse_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
total_clean_mice = clean_mouse_df["Mouse ID"].nunique()
total_clean_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean_regimen = clean_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_regimen = clean_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variance_regimen = clean_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
std_dev_regimen = clean_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
SEM_regimen = clean_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.
summary_df = pd.DataFrame({"Mean Tumor Volume": mean_regimen,
                           "Median Tumor Volume": median_regimen,
                           "Tumor Volume Variance": variance_regimen,
                           "Tumor Volume Std.Dev.": std_dev_regimen,
                           "Tumor Volume Std.Err.": SEM_regimen
                          })
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
aggregation_summary_df = clean_mouse_df.groupby("Drug Regimen")[["Tumor Volume (mm3)"]].agg(["mean", "median", "var", "std", "sem"])
aggregation_summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = merged_df.groupby("Mouse ID").max()
timepoint_reset = last_timepoint.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_timepoint_df = pd.merge(timepoint_reset[["Mouse ID", "Timepoint"]],merged_df, on=["Mouse ID", "Timepoint"], how="left")
merged_timepoint_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
cap_tumor_vol = []
ram_tumor_vol = []
inf_tumor_vol = []
ceft_tumor_vol = []

for treatment in range(len(merged_timepoint_df)-1):
    if merged_timepoint_df.loc[treatment, "Drug Regimen"] == "Capomulin":
        cap_tumor_vol.append(merged_timepoint_df.loc[treatment, "Tumor Volume (mm3)"])
    elif merged_timepoint_df.loc[treatment, "Drug Regimen"] == "Ramicane":
        ram_tumor_vol.append(merged_timepoint_df.loc[treatment, "Tumor Volume (mm3)"])
    elif merged_timepoint_df.loc[treatment, "Drug Regimen"] == "Infubinol":
        inf_tumor_vol.append(merged_timepoint_df.loc[treatment, "Tumor Volume (mm3)"])
    elif merged_timepoint_df.loc[treatment, "Drug Regimen"] == "Ceftamin":
        ceft_tumor_vol.append(merged_timepoint_df.loc[treatment, "Tumor Volume (mm3)"])

tumor_volume = [cap_tumor_vol, ram_tumor_vol, inf_tumor_vol, ceft_tumor_vol]
# Calculate the IQR and quantitatively determine if there are any potential outliers.
def outliers(regimen):
    regimen_data = merged_timepoint_df.loc[merged_timepoint_df["Drug Regimen"] == regimen]["Tumor Volume (mm3)"]
    quartiles = regimen_data.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lowerbound = lowerq - (1.5 * iqr)
    upperbound = upperq + (1.5 * iqr)


    # Determine outliers using upper and lower bounds
    outliers = regimen_data.loc[(regimen_data < lowerbound) | (regimen_data > upperbound)]
    print(f"{regimen}'s potential outliers: {outliers}")
    return regimen_data

capomulin = outliers("Capomulin")
ramicane = outliers("Ramicane")
infubinol = outliers("Infubinol")
ceftamin = outliers("Ceftamin")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_title("Final Tumor Volume (mm3) For Each Regimen")
ax1.set_ylabel("Final Tumor Volume")
ax1.set_xlabel("Regimen/Treatment")
ax1.boxplot(tumor_volume, flierprops = dict(markerfacecolor = "red", marker = "o", markersize = 12))
ax1.set_xticks([1,2,3,4], ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capo_df = merged_df.loc[merged_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")
avg_capo_vol = capo_df["Tumor Volume (mm3)"].mean()
weight = capo_df["Weight (g)"].unique()
fig1, ax1 = plt.subplots()

plt.scatter(weight, avg_capo_vol, color="blue")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
marker_size=12
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
capo_df = merged_df.loc[merged_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")
avg_capo_vol = capo_df["Tumor Volume (mm3)"].mean()
weight = capo_df["Weight (g)"].unique()
weight = weight.astype(float)
c_slope, c_int, c_r, c_p, c_std_err = st.linregress(weight, avg_capo_vol)
c_fit = c_slope * weight + c_int
correlation_coeff = st.pearsonr(weight, avg_capo_vol)
print(f"The correlation between mouse weight and the average tumor volume is: {round(correlation_coeff[0],2)}")
fig1, ax1 = plt.subplots()
plt.scatter(weight, avg_capo_vol, color="blue")
plt.plot(weight,c_fit,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
